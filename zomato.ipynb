# Missing Values Analysis

```python
missing_values_percent = df.isna().sum() / df.shape[0] * 100

missing_values_df = pd.DataFrame({
    "missing values": df.isna().sum(),
    "percentage": missing_values_percent
})

missing_values_df.sort_values(by=["missing values", "percentage"], ascending=False, inplace=True)
missing_values_df["percentage"] = missing_values_df["percentage"].round(2)
print("Missing Values Analysis:".upper())
print("_" * 30)

print(missing_values_df)
```

# Visualizing NAN Values

```python
sns.barplot(x=missing_values_df.index, y=missing_values_df["percentage"],palette="viridis")
plt.title("Percentage of Missing Values per Column", fontsize=16)
plt.xlabel("Columns", fontsize=12)
plt.ylabel("Percentage(%)", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

| % Missing | Suggested Action                        |
| --------- | --------------------------------------- |

| 530%     | Depends  fill if important             |
| > 4050%  | Usually drop, unless its a key feature |
